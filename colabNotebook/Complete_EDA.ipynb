Q1. Calculate the Electricity Bill in Each Row

```python
df1.head()
```

```python
tarriff = pd.read_csv('/content/Tariff_rates.csv')
```

```python
tarriff.head()
```

```python
electricity_bill = pd.merge(df1, tarriff,how = 'inner', on = 'City')
```

```python
electricity_bill['TariffRate'].unique()
```

```python
  electricity_bill['Bill']=electricity_bill['Total Monthly Hours']*electricity_bill['TariffRate']
```

```python
electricity_bill.head()
```

```python
df1['Total Monthly Hours'].isna().sum()
```

Q2. Calculate the Average Monthly Hours of Usage for the Cities in Descending Order

```python
df1['City'].value_counts()
```

```python
avg_monthly_consp= electricity_bill.groupby('City')['Total Monthly Hours'].mean().sort_values(ascending = False)
```

```python
plt.figure(figsize = (12,5), dpi =400)

a = sns.barplot(data=avg_monthly_consp)

a.bar_label(a.containers[0])
```

```python
plt.figure(figsize = (12,5), dpi =400)
sns.barplot(avg_monthly_consp)
```
